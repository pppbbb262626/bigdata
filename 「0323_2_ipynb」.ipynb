```
library("SportsAnalytics")#載入套件
library("ggplot2")
```

```
NBA1920 <- fetch_NBAPlayerStatistics("19-20")#讀入資料
```

```
qplot(ThreesAttempted,TotalPoints,data = NBA1920)
```

```
qplot(ThreesAttempted,TotalPoints,data = NBA1920,color=Position)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data = NBA1920)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data = NBA1920,color=Position)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data = NBA1920,color=Team)
```

```
qplot(ThreesAttempted,TotalPoints,data = NBA1920,color=Team)
```

```
qplot(FieldGoalsAttempted, TotalPoints, data = NBA1920,
 geom = c("point", "smooth"))
```

```
#畫TotalPoints的直⽅圖/ fill = Position 並⽤守備位置Position著⾊
qplot(TotalPoints, data = NBA1920, fill = Position)
```

```
#facets = . ~ Position 用守備位置Position分群畫圖(直向)
qplot(FieldGoalsAttempted,TotalPoints,
  data = NBA1920,
  facets = Position ~ .)
```

```
#facets = . ~ Position 用守備位置Position分群畫圖(橫向)
qplot(FieldGoalsAttempted,TotalPoints,
  data = NBA1920,
  facets = . ~ Position )
```

```
#facets = . ~ Position 用守備位置Position分群畫圖(直向)
#binwidth = 10 每10分一組
qplot(TotalPoints,
  data = NBA1920,
  facets = Position ~ . ,binwidth = 10 )
```

```
#facets = . ~ Position 用守備位置Position分群畫圖(直向)
#binwidth = 100 每100分一組
qplot(TotalPoints,
  data = NBA1920,
  facets = Position ~ . ,binwidth = 100 )
```
